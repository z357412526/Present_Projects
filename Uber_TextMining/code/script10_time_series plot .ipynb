```python
import json
import pandas as pd
import datetime
with open("uber_all.json") as infile:
    tweet_uber = json.load(infile)
```

```python
tweet_uber.keys()
```

```python
tw_text = [a['text'] for a in tweet_uber['statuses']]
```

```python
#['#taxi' in a for a in tw_text]
```

```python
df=pd.DataFrame()
```

```python
date=pd.Series([datetime.datetime.strptime(a['created_at'],'%a %b %d %H:%M:%S %z %Y').strftime('%Y-%m-%d %H:00')
                for a in tweet_uber['statuses']])
```

```python
token_list=['#taxi','#Taxi','#logo','#design','#uberlogo','#SuperBowl50','#SB50','#Paris',
            '#surgepricing',"#SurgePricing","#surge","surge pricing"]
for token in token_list:
    tokenseries=pd.Series([token in a for a in tw_text])
    df[token]=tokenseries
df['date']=date
```

```python
df1=df.groupby(df['date']).sum()
```
